# 2021 Names

```python
import pandas as pd
import numpy as np
import cleandata as cd
```

```python
df = pd.read_csv('../data/Gemeinde_Exekutive - daten.csv')
```

## Clean Data

```python
df = cd.clean(df)

df['Name_cleaned'] =  df['Name_cleaned'].str.strip()

# Lowercase
df['Name_cleaned'] =  df['Name_cleaned'].str.lower()

# Cut Names
df['firstname'] = df['Name_cleaned'].apply(lambda x: x.split(' ')[0])
df['lastname'] = df['Name_cleaned'].apply(lambda x: x.split(' ')[-1])
```

# Vergleich Präsidentinnen und Namen

```python
df[df.Funktion == 'chair'].groupby('Geschlecht').count()
```

```python
print(df[df.Funktion == 'chair'].groupby('firstname').count().sort_values('Name', ascending=False).reset_index()['firstname'].head(5).tolist())
df[df.Funktion == 'chair'].groupby('firstname').count().sort_values('Name', ascending=False)['Gemeinde'].head(5).sum()
```

Es gibt mehr Präsidenten mit Namen ^ als Präsidentinnen insgesamt

## Top Names

```python
# Männer
df[df.Geschlecht == 'm'].groupby('firstname').count().sort_values('Name', ascending=False).head()
```
